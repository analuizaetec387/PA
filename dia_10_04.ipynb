Escrever um programa que ve se o número é par ou impar

```python
# Entrada

numero = int (input("digite um número: "))

#Processamento/Saída

if numero % 2 == 0:

  print (f" O número {numero} é par.")

else:
  print (f"O numero {numero} é ímpar")
```
