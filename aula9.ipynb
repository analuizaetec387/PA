```python
# 1 Contagem simples, o usuario digita um número e o programa imprimi do zero ate o numero

numero = int(input("Digite um número: "))

for i in range(numero + 1):
  if i % 2 == 0:
    print (i)
```

```python
# 2 - Contagem regressiva, o usuario digita um numero e o programa imprimi do numero ate o zero

numero = int(input("Digite um numero"))

for i in range(numero +1):
  print(numero - i)
```

```python
# 3 numeros pares, o usuario digita um numero e o programa imprimi os numeros pares

numero = int(input("Digite um número: "))

for i in range(numero + 1):
  if i % 2 == 0:
    print (i)
```

```python
# 4 Números impares, o usuario digita um numero e o programa imprimi os numeros impares

numero = int(input("Digite um número: "))

for i in range(numero + 1):
  if i % 2 == 1:
    print (i)
```

```python
# 5 Soma, o usuario digita um número e o programa soma os numeros do zero até o número

numero = int(input("Digite um numero"))

for i in range(0,7):
  print(f"{numero} + {i} = {numero * i}")
```

```python
# 6 Tabuada, o usuiário digita um numero e o programa imprimi a tabuada do numero

numero = int(input("Digite um numero"))

for i in range(11):
  print(f"{numero} x {i} = {numero * i}")
```

```python
# 7 Contar letras, o usuario digita uma palavra e o programa conta as letras da palavra

palavra = input("digite uma palavra: ")

print("quantidade de letras:", len(palavra))
```

```python
# 8 Contar vogais, o usuario digita uma palavra e o programa conta as vogais

palavra = input("Digite uma palavra: ")

vogais = 0

for letra in palavra:
  if letra in "aeiouAEIOU":
    vogais +=1

  print(f"quantidade de vogais:", vogais)
```

```python
# 9 Maior numero digitado, o usuario digita quatro numeros e o maior numero digitado

# Entrada

primeiro = int(input("Digite o primeiro número: "))
segundo = int(input("Digite o segundo número: "))
terceiro = int(input("Digite o terceiro número: "))
quarto = int(input("Digite o quarto número: "))

# Processamento
maior = primeiro
if segundo > maior:
 maior = segundo
 if terceiro > maior:
  maior = terceiro
  if quarto > maior:
     maior = quarto

  print(f"O maior número digitado foi {maior}.")
```

```python
# 10 O usuario digita um numero entre 01 e 100 e o programa mostra quais são divisiveis por 3 e 5

numero = int(input("Digite um número entre 1 e 100: "))

print(f"número divisiveis por 3 e 5 entre 1 e {numero}:")

for i in range(1, numero + 1):
  if i % 3 ==0 and i % 5 ==0:
    print(i)
```
