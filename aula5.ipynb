# Exercício Celsius para Fahrenheit

```python
# Calcular temperatura de Celsius para Fahrenheit

# Entrada

celsius = float(input("Digite a temperatura em celsius:"))

# Processamento
fahrenheit = (celsius * 9/5) + 32

# Saida
print (f"{celsius}°C é igual a {fahrenheit}°F")
```

```python
# Calculadora simples

# Entrada 2 valores
# Processamento calcular usando operadores aritiméticos

# Saída Mostrar o resultado

# Entrada

numero1 = int (input("digite um número: "))
numero2 = int (input("digite o segundo número: "))

# Processamento

soma = numero1 + numero2
subtração = numero1 - numero2
multiplicação = numero1 * numero2
divisão = numero1 / numero2
div_inteira = numero1 // numero2
resto = numero1 % numero2
potência = numero1 ** numero2

# Saída

print (f"o resultado da soma de {numero1} e {numero2} é igual: {soma}")
print (f"o resultado da subtração de {numero1} e {numero2} é igual: {subtração}")
print (f"o resultado da multiplicação de {numero1} e {numero2} é igual: {multiplicação}")
print (f"o resultado da divisão de {numero1} e {numero2} é igual: {divisão}")
print (f"o resultado da divisão inteira de {numero1} e {numero2} é igual: {div_inteira}")
print (f"o resultado da divisão de {numero1} e {numero2} é igual: {resto}")
print (f"o resultado da potência de {numero1} e {numero2} é igual: {potência}")

```
