```python
# Peça a nota de um aluno e informe:
# Média menor que 5 - Reprovado
# Média maior que 5 e menor que 7 - Recuperação
# Média maior ou igual a 7 - Aprovado

# Entrada

nota1 = float (input("Digite a nota do primeiro bimestre: "))

if nota1 < 0 or nota1 > 10:
  print ("nota inválida, digite uma nota de 0 a 10:")
  nota1 = float (input("Digite a nota do primeiro bimestre: "))

nota2 = float (input("Digite a nota do segundo bimestre: "))

if nota2 < 0 or nota1 > 10:
  print ("nota inválida, digite uma nota de 0 a 10:")
  nota2 = float (input("Digite a nota do segundo bimestre: "))

nota3 = float (input("Digite a nota do terceiro bimestre: "))

if nota3 < 0 or nota1 > 10:
  print ("nota inválida, digite uma nota de 0 a 10:")
  nota2 = float (input("Digite a nota do terceiro bimestre: "))

nota4 = float (input("Digite a nota do quarto bimestre: "))

if nota4 < 0 or nota1 > 10:
  print ("nota inválida, digite uma nota de 0 a 10:")
  nota4 = float (input("Digite a nota do quarto bimestre: "))

# Processamento

media = (nota1 + nota2 + nota3 + nota4)/4

if media < 5:
 print (f"reprovado, sua média é {media}")

elif media <= 7:
  print (f"recuperação, sua média é {media}")

else:
  print (f"aprovado, sua média é {media}")
```

```python
# Dia da semana

# Entrada

dia = int(input("Digite um numero: "))

# Processamento / saida

if dia == 1:
  print("Domingo")

elif dia == 2:
  print("Segunda-feira")

elif dia == 3:
  print("Terça-feira")

elif dia == 4:
  print("Quarta-feira")

elif dia == 5:
  print("Quinta-feira")

elif dia == 6:
  print("Sexta-feira")

elif dia == 7:
  print("Sabado")

else:
  print("Numero inválido")
```
